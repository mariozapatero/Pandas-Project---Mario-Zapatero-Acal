# ETAPA DE VISUALIZACIÓN

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import plotly.express as px
```

```python
attacks = pd.read_csv("data/attacks_clean_def.csv",encoding = 'UTF-8')
```

## Hipótesis 1

> "*Los surfistas son la principal víctima de los tiburones.*"

```python
fig = px.histogram(attacks, x="activity",
                  title='Frecuencia por Actividad',
                  category_orders={"activity":["surfing","swimming","fishing", "diving", "shark_activities", "other"]})
fig.update_layout(bargap=0.2)
fig.show()
```

```python
sns.countplot(x=attacks.activity);
```

```python
fig = px.histogram(attacks, x="activity", histnorm='percent',
                  title='Porcentaje por Actividad',
                   color_discrete_sequence=['#330C73'],
                   opacity=0.75,
                   category_orders={"activity":["surfing","swimming","fishing", "diving", "shark_activities", "other"]},
                   )
fig.update_layout(bargap=0.2)
fig.update_xaxes(categoryorder='array')
fig.show()

# ¿Cuál es el porcentaje de ataques para cada actividad (sobre el total de ataques)?
```

Es más interesante para el análisis fijarnos únicamente en las 5 actividades principales.

```python
fig = px.histogram(attacks, x="activity", title='Hacemos zoom para descartar "other":')
fig.update_layout(bargap=0.2)
fig.show()

# Hacemos zoom descartando "other"
```

Observamos como "surfing" es la actividad con mayor frecuencia en los ataques de tiburones.

De hecho, ni siquiera la actividad "other" que engloba gran cantidad de actividades distintas, llega a superar la frecuencia de "surfing".

Los ataques a surfistas representan más de un 25% sobre el total de ataques de tiburones en el mundo.

Si nos centramos en las 5 principales actividades consideradas, los ataques a surfistas representan el 33,32% sobre el subtotal.

Además, comprobamos que la frecuencia de surfing es muy superior a la de las otras actividades, valores que podemos consultar fácilmente gracias a los gráficos interactivos de plotly (podríamos tratar de añadir etiquetas de datos, pero no tendría demasiado sentido ya que esta es una de las principales ventajas que ofrece plotly).

**Resultado: la hipótesis SÍ se cumple.**

## Hipótesis 2

> "*La mayoría de los ataques de tiburón en el mundo son 'fatal attacks'.*"

```python
fig = px.histogram(attacks, x="fatal_y_or_n", color_discrete_sequence=['#90EE90'], title='Frecuencia de Muerte')
fig.update_layout(bargap=0.2)
fig.show()
```

Observamos rápidamente como la hipótesis no se cumple (la mayoría de los ataques de tiburón no son letales).

```python
fig = px.histogram(attacks, x="fatal_y_or_n", color_discrete_sequence=['#66CDAA'], histnorm='percent', title='Tasa de Mortalidad')
       
fig.update_layout(bargap=0.2)
fig.show()
```

Tan solo un 22,87% de los ataques de tiburón provocan la muerte en las víctimas, teniendo en cuenta que desconocemos las consecuencias para un 8,87% de los ataques.

Descartando este grupo de ataques de consecuencias desconocidas, podemos estimar que la tasa de mortalidad de los ataques de tiburón a nivel mundial a penas supera el 26,3%.

**Resultado: la hipótesis NO se cumple.**
