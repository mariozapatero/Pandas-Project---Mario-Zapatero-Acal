# Limpieza

Empezamos gestionando los null encontrados en la exploración inicial.

   1. Eliminamos columnas "Unnamed: 22" y "Unnamed: 23".

```python
attacks.drop(["Unnamed: 22", "Unnamed: 23"], axis=1, inplace=True)
```

   2. Eliminamos los registros en los que todos los valores son nulos. 

```python
attacks.dropna(how="all", inplace=True)
```

   3. Eliminamos todos los registros en los que todos los valores sean NaN, excepto el Case Number (no aporta valor).

   3. 1. Eliminamos todos los registros en los que el Case Number sea igual a 0 (porque en principio todos los registros con Case Number = 0 tienen NaN en el resto de valores).

```python
attacks.loc[(attacks["Case Number"] == "0")]
```

```python
attacks.drop(attacks[(attacks["Case Number"] == "0")].index, inplace=True)
```

```python
attacks.tail()
```

   3. 2. Eliminamos el registro con "xx" en la columna "Case Number" porque observamos que el resto de valores son NaN.

```python
attacks.loc[(attacks["Case Number"] == "xx")]
```

```python
attacks.drop(attacks[(attacks["Case Number"] == "xx")].index, inplace=True)
```

```python
attacks.tail()
```

```python
attacks["case_number"].equals(attacks["case_number.1"])
```

```python
attacks["case_number"].equals(attacks["case_number.2"])
```

```python
attacks["case_number.1"].equals(attacks["case_number.2"])
```

   Observamos que esta hipótesis no se cumple.

No nos aportan información adicional, por lo que decidimos eliminar las columnas (no podremos hacerlo a través de métodos relacionados con duplicates porque no son duplicados, así que lo haremos mediante .drop
