Para empezar, establecemos las siguientes categorÃ­as y coincidencia bÃ¡sicas para cada categorÃ­a:

    - Surfing = todo lo que contenga "surf", "wave" o "board" (todo lo que sea con una tabla o relacionado con las olas, sin distinguir entre disciplinas o tipos de surf).
    
    - Swimming = todo lo que contenga "swim" o "snorkel".
    
    - Diving = todo lo que contenga "diving" (sin distinguir el tipo de buceo).
    
    - Fishing = todo lo que contenga "fish".
       
    - Shark Activities = todo lo que contenga "sharks" (relacionado con los tiburones).
    
    - Saling = todo lo que contenga "boat", "ship", "ferry", "raft", "rowing", "paddling", "kayak" (sin distinguir entre embarcaciones).

Generaremos primero estas 5 categorÃ­as bÃ¡sicas agrupadas y valoraremos la situaciÃ³n despuÃ©s de hacerlo.

El orden de ejecutar el .replace serÃ¡ importante ya que algunos valores comparten palabras clave de distintas categorÃ­as ("swim" y "fish", por ejemplo) y el replace definitivo serÃ¡ el primero que se ejcute.

```python
def activities (i):
    if 'surf' in str(i).lower():
        return 'surfing'
    
    elif "wave" in str(i).lower():
        return 'surfing'
    
    elif "board" in str(i).lower():
        return 'surfing'
    
    
    elif "swim" in str(i).lower():
        return 'swimming'
    
    elif "snorkel" in str(i).lower():
        return 'swimming'
    
    elif "bath" in str(i).lower():
        return 'swimming'
    
    
    elif "diving" in str(i).lower():
        return 'diving'
    
    
    elif "fish" in str(i).lower():
        return 'fishing'
    
    
    elif "shark" in str(i).lower():
        return 'shark_activities'
    
    
    elif "boat" in str(i).lower():
        return 'sailing'
    
    elif "ship" in str(i).lower():
        return 'sailing'
    
    elif "ferry" in str(i).lower():
        return 'sailing'
    
    elif "raft" in str(i).lower():
        return 'sailing'
    
    elif "rowing" in str(i).lower():
        return 'sailing'
    
    elif "paddling" in str(i).lower():
        return 'sailing'
    
    elif "kayak" in str(i).lower():
        return 'sailing'
    
    
    else:
        return str(i).lower()
    
```

```python
attacks["activity"] = attacks["activity"].apply(activities)
```

Observamos como la cantidad de valores Ãºnicos se ha reducido a menos de un cuarto (hemos limpiado casi 1.200 valores asignÃ¡ndolos a una nueva categorÃ­a), lo cual supone un avance importante en la limpieza de los datos de la columna "activity".

Tras valorar la situaciÃ³n, decidimos incluir todos estos valores en una nueva categorÃ­a llamada "other" que englobe a todas las actividades no incluÃ­das en las otras categorÃ­as bÃ¡sicas.

Volvemos a definir la funciÃ³n activities() incluyendo esta nueva consideraciÃ³n:

  7. Procedemos a limpiar la columna **"fatal_y/n"**

Nos damos cuenta de que el caracter "/" da problemas a la hora de trabajar con el nombre de la columna.

Cambiamos el nombre para arreglarlo:

```python
rename = {"fatal_y/n" : "fatal_y_or_n"}

attacks.rename(columns = rename, inplace=True)
```

```python
attacks.sample()
```

Extraemos los valores Ãºnicos en esta columna y observamos que existen diferentes posibles valores (y entre ellos valores NaN). 

Procedemos a arreglar este problema:

```python
list(attacks.fatal_y_or_n.unique())
```

Primero procedemos a estandarizar los values con errores de escritura a "N" o "Y".

```python
def standar (i):
    if 'N' in str(i).upper() and "UNKNOWN" not in str(i).upper() and "NAN" not in str(i).upper():
        return 'N'
    
    elif 'Y' in str(i).upper():
        return "Y"
    
    else:
        return str(i).upper()
```

```python
attacks["fatal_y_or_n"] = attacks["fatal_y_or_n"].apply(standar)
```

Extraemos los valores Ãºnicos que tenemos ahora:

```python
list(attacks.fatal_y_or_n.unique())
```

Hemos resucido los valores Ãºnicos al corregir los errores de escritura.

Ahora tenemos los siguientes inconvenientes:

    - Existe al menos un valor "2017" y otro valor "M".
    
    - Los valores NaN son ahora "NAN" por la ejecuciÃ³n de la funciÃ³n anterior. Esto puede ser un inconveniente, pero como vamos a querer sustituirlos por otro valor, esto no deberÃ­a influir en la limpieza (esto en realidad ha sido un error en la funciÃ³n, pero trataremos de seguir trabajando a pesar de ello (tambiÃ©n podrÃ­amos volver a ejecutar el cÃ³digo)).

Comprobamos la cantidad de datos que existen para cada valor:

```python
attacks["fatal_y_or_n"].value_counts()
```

Como solo existe un "2017" y un "M", decidimos incluirlos en "UNKNOWN" (podrÃ­amos deducir que "M" se refiere a "N", pero al ser un Ãºnico valor quizÃ¡ sea mejor no tener esto en cuenta y tratar de evitar interpretaciones personales).

```python
def unknown (i):
    if i == "M":
        return "UNKNOWN"
    elif i == "2017":
        return "UNKNOWN"
    else:
        return i
```

```python
attacks["fatal_y_or_n"] = attacks["fatal_y_or_n"].apply(unknown)
```

Ahora ya solo nos queda el problema de los "NAN".

En esta columna solo hay dos posibles valores: fatal attack = yes (Y) o fatal attack = no (N).

No tiene sentido rellenar estos valores nulos con "Y" o "N" segÃºn lo consideremos. Sin embargo, nos damos cuenta de que en todos aquellos registros en los que el valor de la columna "injury" contiene "fatal", la columna "fatal_y_or_n" contiene el valor "Y". Este es el Ãºnico criterio que podemos seguir, ya que no podemos entrar a valorar la gravedad del ataque segÃºn la informaciÃ³n recogida en "injury".

DecisiÃ³n --> rellenar los NaN con:
   - "Y" si el valor de "injury" contiene "fatal".
   - "UNKNOWN" si el valor de "injury" no contiene "fatal".

```python
attacks["fatal_y_or_n"].value_counts()
```

```python
for i,row in attacks.iterrows(): # Muy interesante .iterrows() para ir iterando por todas los valores de cada fila.
    if ("fatal" in str(row["injury"]).lower()) & (row["fatal_y_or_n"]=="NAN"): # Si se cumplen las dos condiciones que queremos ("injury" contiene "fatal" y "fatal_y_or_n" es igual a "NAN")
        attacks.loc[i, "fatal_y_or_n" ] = "Y" # Sustituimos el valor en "fatal_y_or_n" ("NAN") por "Y" (esta sintaxis la hemos obtenido a travÃ©s las sugerencias que nos daban en un error que hemos obtenido previamente).
        
    else:
        pass # Si las condiciones no se cumplen queremos que todo se mantenga como estÃ¡.
```

Observamos como la cantidad de "NAN" ha reducido. Esto indica que hemos conseguido lo que querÃ­amos (sustituir los valores "NAN" por "Y" si la columna "injury" contiene "fatal".

```python
attacks["fatal_y_or_n"].value_counts()
```

Ahora cambiaremos el resto de valores "NAN" por "UNKNOWN" porque no podemos garantizar que no contaminamos los datos si los incluimos en alguna otra categorÃ­a (no tenemos informaciÃ³n veraz para incluirlos en "Y" o "N").

```python
def unknown2 (i):
    if i == "NAN":
        return "UNKNOWN"
    else:
        return i
```
